```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('tiempos extraccion.xlsx', sheet_name='ANEXO')
df.sample(n=5)
```

```python
# delete unnecesary columns
df.drop(columns='Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
# Pivot table
df.pivot_table(index='RUTA', columns='EQ', values='TT')
```

```python
df.pivot_table(index='EQ', columns='RUTA', values='TC')
```

```python
df[['EQ','EQC', 'TC']].groupby(['EQ','EQC']).mean()
```

```python
def tipo_eq(eq):
    if eq.startswith('D-'):
        return 'Dumper'
    if eq.startswith('VQ-'):
        return 'Volquete'
```

```python
# Tipo de equpo TEQ
df['TEQ'] = df['EQ'].apply(tipo_eq)
```

```python
df[['TEQ','EQC', 'TC']].groupby(['TEQ','EQC']).mean()
```

```python
df.pivot_table(index='TEQ', columns='EQC', values='TC')
```
